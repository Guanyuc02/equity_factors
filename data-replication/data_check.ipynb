```python
# Base NYSE holiday rules
class _NYSEBaseCalendar(AbstractHolidayCalendar):
    rules = [
        Holiday("New Year's Day", month=1, day=1, observance=sunday_to_monday),
        Holiday("Martin Luther King Jr. Day", month=1, day=1, offset=DateOffset(weekday=MO(3)), start_date="1998-01-01"),
        Holiday("Presidents Day", month=2, day=1, offset=DateOffset(weekday=MO(3))),
        Holiday("Good Friday", month=1, day=1, offset=[Easter(), Day(-2)]),
        Holiday("Memorial Day", month=5, day=31, offset=DateOffset(weekday=MO(-1))),
        Holiday("Juneteenth", month=6, day=19, observance=nearest_workday, start_date="2021-06-01"),
        Holiday("Independence Day", month=7, day=4, observance=nearest_workday),
        Holiday("Labor Day", month=9, day=1, offset=DateOffset(weekday=MO(1))),
        Holiday("Thanksgiving", month=11, day=1, offset=DateOffset(weekday=TH(4))),
        Holiday("Christmas Day", month=12, day=25, observance=nearest_workday),
    ]

# Extend with ad-hoc full-day closures that affect month ends
class NYSEWithAdhoc(_NYSEBaseCalendar):
    def __init__(self):
        super().__init__()
        self.adhoc = [
            pd.Timestamp("2001-09-11"), pd.Timestamp("2001-09-12"),
            pd.Timestamp("2001-09-13"), pd.Timestamp("2001-09-14"),
            pd.Timestamp("2004-06-11"),  # Reagan funeral
            pd.Timestamp("2007-01-02"),  # Ford funeral
            pd.Timestamp("2012-10-29"), pd.Timestamp("2012-10-30"),  # Sandy
            pd.Timestamp("2018-12-05"),  # G.H.W. Bush funeral
        ]

# Custom business day: NYSE trading days
_NYSE_BDAY = CustomBusinessDay(calendar=NYSEWithAdhoc())
```

```python
def _to_month_last_weekday(s: pd.Series) -> pd.Series:
    """
    Map input dates (YYYYMMDD, YYYYMM, or parseable) to the last NYSE trading DAY of that month.
    Handles weekend roll and U.S. market holidays (incl. Good Friday, Memorial Day, Juneteenth, etc.),
    plus major ad-hoc closures listed above.
    """
    s = s.astype(str).str.strip()

    if s.str.fullmatch(r"\d{8}").all():
        dt = pd.to_datetime(s, format="%Y%m%d", errors="coerce")
    elif s.str.fullmatch(r"\d{6}").all():
        dt = pd.to_datetime(s, format="%Y%m", errors="coerce")
    else:
        dt = pd.to_datetime(s, errors="coerce")

    # Convert to month periods
    per = dt.dt.to_period("M")
    uniq = per.dropna().unique()

    # For each month, compute the last NYSE trading day
    last_trading_by_month = {}
    for p in uniq:
        start = p.to_timestamp(how="start")
        end = p.to_timestamp(how="end")
        # All NYSE business days in the month
        bdays = pd.bdate_range(start, end, freq=_NYSE_BDAY)
        if len(bdays) == 0:
            # Fallback: if no trading days (should not happen), use calendar month-end
            last_trading_by_month[str(p)] = end.normalize()
        else:
            last_trading_by_month[str(p)] = bdays[-1].normalize()

    out = per.astype(str).map(last_trading_by_month)
    return pd.to_datetime(out)
```

```python
def align_monthly_and_compare(csv1_path: str, csv2_path: str, save_path: str | None = None) -> pd.DataFrame:
    """
    1) Load FF5, extract the table, convert Date to month-end.
    2) Load factors_copy, convert (ensure) Date to month-end.
    3) Filter FF5 so its first row's date >= factors_copy's first row's date.
    4) For shared columns, compute delta = FF5 - factors_copy by month and report min, max, mean.
    5) Save the filtered FF5 as a new CSV. Return the summary DataFrame.
    """
    with open(csv1_path, "r", encoding="utf-8", errors="ignore") as f:
        lines = f.read().splitlines()
    header_idx = None
    for i, ln in enumerate(lines):
        low = ln.lower()
        if "," in ln and ("date" in low) and (("mkt-rf" in low) or ("mktrf" in low)) and ("smb" in low) and ("hml" in low):
            header_idx = i
            break
    if header_idx is None:
        raise ValueError("Could not locate Fama-French monthly header in csv1.")
    data_lines = [lines[header_idx]]
    for ln in lines[header_idx + 1:]:
        if not ln.strip():
            break
        if not (ln[0].isdigit() and "," in ln):
            break
        data_lines.append(ln)
    df1 = pd.read_csv(StringIO("\n".join(data_lines)))
    if "Date" not in df1.columns:
        raise ValueError("FF5 monthly table lacks 'Date' column after parsing.")

    df2 = pd.read_csv(csv2_path)
    if "Date" not in df2.columns:
        raise ValueError("factors_copy must contain a 'Date' column.")

    df1["Date"] = _to_month_last_weekday(df1["Date"])
    df2["Date"] = _to_month_last_weekday(df2["Date"])
    df1 = df1.dropna(subset=["Date"]).sort_values("Date").reset_index(drop=True)
    df2 = df2.dropna(subset=["Date"]).sort_values("Date").reset_index(drop=True)
    if df2.empty:
        raise ValueError("csv2 has no parsable monthly dates.")
    start_date = df2["Date"].iloc[0]
    df1_aligned = df1[df1["Date"] >= start_date].copy()

    num_cols = [c for c in df1_aligned.columns if c != "Date"]
    for c in num_cols:
        df1_aligned[c] = pd.to_numeric(df1_aligned[c], errors="coerce") / 100.0

    if "Mkt-RF" in df1_aligned.columns:
        df1_aligned = df1_aligned.rename(columns={"Mkt-RF": "MktRf"})

    # --- normalize column names to match across hyphen/case variants ---
    def normcols(cols):
        return {c: re.sub(r"[^0-9a-z]+", "", c.lower()) for c in cols}
    nmap1 = normcols(df1_aligned.columns)
    nmap2 = normcols(df2.columns)
    d1 = df1_aligned.rename(columns=nmap1)
    d2 = df2.rename(columns=nmap2)

    # --- choose shared numeric columns excluding date ---
    shared = sorted(set(d1.columns).intersection(d2.columns) - {"date"})
    for c in shared:
        d1[c] = pd.to_numeric(d1[c], errors="coerce")
        d2[c] = pd.to_numeric(d2[c], errors="coerce")
    shared = [c for c in shared if pd.api.types.is_numeric_dtype(d1[c]) and pd.api.types.is_numeric_dtype(d2[c])]
    if not shared:
        raise ValueError("No shared numeric columns to compare.")

    # --- merge on month-end Date and compute deltas ---
    merged = d1.merge(d2[["date"] + shared], how="inner", left_on="date", right_on="date", suffixes=("_1", "_2"))
    if merged.empty:
        raise ValueError("No overlapping months after alignment.")

    out_dir = Path("./cleaned_public/FF5_aligned_plots")
    out_dir.mkdir(parents=True, exist_ok=True)

    rows = []
    for c in shared:
        delta = merged[f"{c}_1"] - merged[f"{c}_2"]
        rows.append({
            "column": c,                
            "rows_compared": int(delta.shape[0]),
            "delta_min": float(np.nanmin(delta)),
            "delta_max": float(np.nanmax(delta)),
            "delta_mean": float(np.nanmean(delta)),
        })
        plt.figure()
        plt.plot(merged["date"], delta)
        plt.title(f"Delta over time: {c}")
        plt.xlabel("Date")
        plt.ylabel("Delta")
        plt.tight_layout()
        out_png = Path(out_dir) / f"delta_{c}.png"
        plt.savefig(out_png)
        plt.close()
        
    summary = pd.DataFrame(rows).sort_values("column").reset_index(drop=True)

    # --- save filtered original csv1 with original column names ---
    if save_path is None:
        save_path = str("./cleaned_public/FF5_aligned.csv")
    df1_aligned.to_csv(save_path, index=False)

    return summary

align_monthly_and_compare('./public_source/FF5.csv', 'factors_copy.csv')
```

Main changes: 
- CRSP CIZ methodology change (2025): monthly stock returns now equal compounded daily returns with dividends on ex-dates. All factor legs built from CRSP returns move.
- Full-history rebuilds each update: French reconstructs the entire history when CRSP backfills or corrects data. Portfolio constituents and returns shift, so historical factor values change.


Documented rule/definition tweaks over time that alter sorts and portfolio legs:
- Book equity adjustments: FASB-109 change (2016) and removal of the FASB-106 adjustment (2020) affect B/M sorts, hence HML.
- Operating profitability definition updated in 2018 to include minority interest in the denominator, moving RMW.
- CRSP–Compustat linking moved to CRSP’s links in 2021, which can reassign firm fundamentals and group membership for HML/RMW/CMA/SMB.
- Risk-free series source change (2024-06): RF moved from Ibbotson to ICE BofA 1-Month T-Bill Index, affecting Mkt-RF deltas but not SMB/HML/RMW/CMA.
- CRSP data program changes and corrections (shares outstanding, delists, etc.) that ripple into factor legs; Fama–French quantify the impact of process changes plus five major CRSP projects on SMB/HML means.

```python
def _norm(s: str) -> str:
    return "".join(ch for ch in s.lower() if ch.isalnum()) 

def _find_col(df: pd.DataFrame, candidates: List[str]) -> Optional[str]:
    nmap = {_norm(c): c for c in df.columns}
    for cand in candidates:
        key = _norm(cand)
        if key in nmap:
            return nmap[key]
    return None

def plot_ff5_log_cum_returns(
    aligned_ff5_csv: str = "./cleaned_public/FF5_aligned.csv",
    factors_copy_csv: str = "factors_copy.csv",
    out_dir: str = "./cleaned_public/FF5_aligned_plots",
) -> None:
    outp = Path(out_dir)
    outp.mkdir(parents=True, exist_ok=True)

    a = pd.read_csv(aligned_ff5_csv)
    f = pd.read_csv(factors_copy_csv)

    for df in (a, f):
        date_col = next((c for c in df.columns if c.lower() == "date"), None)
        if date_col is None:
            raise ValueError("Both datasets must contain a 'Date' column.")
        if date_col != "Date":
            df.rename(columns={date_col: "Date"}, inplace=True)
        df["Date"] = _to_month_last_weekday(df["Date"])

    factors: Dict[str, List[str]] = {
        "MktRf": ["MktRf", "Mkt-RF", "MKTRF", "MKT_RF", "mkt_rf"],
        "SMB":   ["SMB"],
        "HML":   ["HML"],
        "RMW":   ["RMW"],
        "CMA":   ["CMA"],
        "RF":    ["RF", "Rf", "rf"],
    }

    # Find matching columns in both frames
    found = []
    for canon, aliases in factors.items():
        ca = _find_col(a, aliases)
        cf = _find_col(f, aliases)
        if ca and cf:
            found.append((canon, ca, cf))

    if not found:
        raise ValueError("No overlapping FF5 factor columns found between aligned FF5 and factors_copy.")

    # Join on dates once
    a_small = a[["Date"] + [ca for _, ca, _ in found]].copy()
    f_small = f[["Date"] + [cf for _, _, cf in found]].copy()
    m = a_small.merge(f_small, on="Date", how="inner", suffixes=("_aligned", "_copy"))
    m = m.sort_values("Date").reset_index(drop=True)

    # Coerce numeric
    for _, ca, cf in found:
        m[ca + "_aligned"] = pd.to_numeric(m[ca + "_aligned"], errors="coerce")
        m[cf + "_copy"] = pd.to_numeric(m[cf + "_copy"], errors="coerce")

    # Plot each factor
    for canon, ca, cf in found:
        tmp = m[["Date", ca + "_aligned", cf + "_copy"]].replace([np.inf, -np.inf], np.nan).dropna()
        tmp = tmp[(tmp[ca + "_aligned"] > -1.0) & (tmp[cf + "_copy"] > -1.0)].copy()
        if tmp.empty:
            continue

        tmp["log_cum_aligned"] = np.log1p(tmp[ca + "_aligned"]).cumsum()
        tmp["log_cum_copy"] = np.log1p(tmp[cf + "_copy"]).cumsum()

        plt.figure()
        plt.plot(tmp["Date"], tmp["log_cum_aligned"], label=f"{canon} (aligned)")
        plt.plot(tmp["Date"], tmp["log_cum_copy"], label=f"{canon} (factors_copy)")
        plt.title(f"Log cumulative return: {canon}")
        plt.xlabel("Date")
        plt.ylabel("log cum return")
        plt.legend()
        plt.tight_layout()
        plt.savefig(outp / f"log_cum_{canon}.png")
        plt.close()

plot_ff5_log_cum_returns(
     aligned_ff5_csv="./cleaned_public/FF5_aligned.csv",
     factors_copy_csv="factors_copy.csv",
)
```

```python
def align_liquidity_and_compare(
    liq_txt_path: str,
    factors_copy_csv_path: str,
    save_dir: Optional[str] = None,
    save_filtered_csv: Optional[str] = None,
) -> pd.DataFrame:
    def to_month_end(s: pd.Series) -> pd.Series:
        s = s.astype(str).str.strip()
        if s.str.fullmatch(r"\d{8}").all():
            dt = pd.to_datetime(s, format="%Y%m%d", errors="coerce")
        elif s.str.fullmatch(r"\d{6}").all():
            dt = pd.to_datetime(s, format="%Y%m", errors="coerce")
        else:
            dt = pd.to_datetime(s, errors="coerce", infer_datetime_format=True)
        return dt.dt.to_period("M").dt.to_timestamp("M")

    liq_cols = ["Month", "Agg Liq.", "Innov Liq (eq8)", "Traded Liq (LIQ_V)"]
    df_liq = pd.read_csv(
        liq_txt_path, sep=r"\s+", engine="python", comment="%", header=None, names=liq_cols, dtype={"Month": str}
    )
    df_liq = df_liq.rename(columns={"Month": "Date"})
    df_liq["Date"] = to_month_end(df_liq["Date"])
    for c in ["Agg Liq.", "Innov Liq (eq8)", "Traded Liq (LIQ_V)"]:
        df_liq[c] = pd.to_numeric(df_liq[c], errors="coerce").replace(-99, np.nan)
    df_liq = df_liq.dropna(subset=["Date"]).sort_values("Date").reset_index(drop=True)

    df_fc = pd.read_csv(factors_copy_csv_path)
    date_col = next((c for c in df_fc.columns if c.lower() == "date"), None)
    if date_col is None:
        raise ValueError("factors_copy must contain a 'Date' column.")
    if date_col != "Date":
        df_fc = df_fc.rename(columns={date_col: "Date"})
    if "LIQ_PS" not in df_fc.columns:
        raise ValueError("factors_copy must contain 'LIQ_PS'.")

    df_liq["Date"] = _to_month_last_weekday(df_liq["Date"])
    df_fc["Date"]  = _to_month_last_weekday(df_fc["Date"])
    df_fc = df_fc.dropna(subset=["Date"]).sort_values("Date").reset_index(drop=True)

    start_date = df_fc["Date"].iloc[0]
    df_liq_aligned = df_liq[df_liq["Date"] >= start_date].copy()
    merged = df_liq_aligned.merge(df_fc[["Date", "LIQ_PS"]], on="Date", how="inner")
    if merged.empty:
        raise ValueError("No overlapping months between LIQ and factors_copy.")

    merged["delta_LIQ_V_minus_LIQ_PS"] = merged["Traded Liq (LIQ_V)"] - merged["LIQ_PS"]
    delta = merged["delta_LIQ_V_minus_LIQ_PS"].to_numpy()
    summary = pd.DataFrame(
        [{
            "column": "LIQ_V_vs_LIQ_PS",
            "rows_compared": int(np.sum(~np.isnan(delta))),
            "delta_min": float(np.nanmin(delta)),
            "delta_max": float(np.nanmax(delta)),
            "delta_mean": float(np.nanmean(delta)),
        }]
    )

    out_dir = Path(save_dir or "./cleaned_public/liquidity_deltas")
    out_dir.mkdir(parents=True, exist_ok=True)
    plt.figure()
    plt.plot(merged["Date"], merged["delta_LIQ_V_minus_LIQ_PS"])
    plt.title("Delta over time: LIQ_V - LIQ_PS")
    plt.xlabel("Date")
    plt.ylabel("Delta")
    plt.tight_layout()
    plt.savefig(out_dir / "delta_LIQ_V_minus_LIQ_PS.png")
    plt.close()

    filtered_csv = Path(save_filtered_csv or "./cleaned_public/LIQ_aligned.csv")
    filtered_csv.parent.mkdir(parents=True, exist_ok=True)
    df_liq_to_save = (
        df_liq_aligned[["Date", "Traded Liq (LIQ_V)"]]
        .rename(columns={"Traded Liq (LIQ_V)": "LIQ_PS"})
    )
    df_liq_to_save.to_csv(filtered_csv, index=False)

    return summary
align_liquidity_and_compare(
    liq_txt_path="./public_source/liq_pq.txt",
    factors_copy_csv_path="factors_copy.csv",
)
```

Main Changes:
- CRSP backfills/corrections. Pástor and Stambaugh note that CRSP performs corrections to its historical data, so a dataset downloaded years later is not identical even over the same span. Replication differences versus their website series were traced to these year-to-year CRSP updates.
- Ongoing author updates. The authors continue to update the file through recent months, so the distributed history reflects the latest CRSP vintage and any implementation refinements.

```python
def plot_liq_ps_log_cum_return(
    aligned_csv: str = "./cleaned_public/LIQ_aligned.csv",
    factors_copy_csv: str = "factors_copy.csv",
    out_dir: str = "./cleaned_public/liquidity_deltas",
) -> None:
    outp = Path(out_dir)
    outp.mkdir(parents=True, exist_ok=True)

    a = pd.read_csv(aligned_csv)
    if "Date" not in a or "LIQ_PS" not in a:
        raise ValueError("Aligned file must have columns: Date, LIQ_PS.")
    a["Date"] = _to_month_last_weekday(a["Date"])
    a["LIQ_PS"] = pd.to_numeric(a["LIQ_PS"], errors="coerce")

    f = pd.read_csv(factors_copy_csv)
    date_col = next((c for c in f.columns if c.lower() == "date"), None)
    if date_col is None or "LIQ_PS" not in f.columns:
        raise ValueError("factors_copy must have Date and LIQ_PS.")
    if date_col != "Date":
        f = f.rename(columns={date_col: "Date"})
    f["Date"] = _to_month_last_weekday(f["Date"])
    f["LIQ_PS"] = pd.to_numeric(f["LIQ_PS"], errors="coerce")

    m = (
        a[["Date", "LIQ_PS"]].rename(columns={"LIQ_PS": "LIQ_PS_aligned"})
        .merge(f[["Date", "LIQ_PS"]].rename(columns={"LIQ_PS": "LIQ_PS_copy"}),
               on="Date", how="inner")
        .sort_values("Date")
        .reset_index(drop=True)
    )

    m = m.replace([np.inf, -np.inf], np.nan).dropna(subset=["LIQ_PS_aligned", "LIQ_PS_copy"])
    m = m[(m["LIQ_PS_aligned"] > -1.0) & (m["LIQ_PS_copy"] > -1.0)]

    m["log_cum_aligned"] = np.log1p(m["LIQ_PS_aligned"]).cumsum()
    m["log_cum_copy"] = np.log1p(m["LIQ_PS_copy"]).cumsum()

    plt.figure()
    plt.plot(m["Date"], m["log_cum_aligned"], label="LIQ_PS (aligned)")
    plt.plot(m["Date"], m["log_cum_copy"], label="LIQ_PS (factors_copy)")
    plt.title("Log cumulative return of LIQ_PS")
    plt.xlabel("Date")
    plt.ylabel("log cum return")
    plt.legend()
    plt.tight_layout()
    plt.savefig(outp / "liq_ps_log_cum_return.png")
    plt.close()

plot_liq_ps_log_cum_return(
    aligned_csv="./cleaned_public/LIQ_aligned.csv",
    factors_copy_csv="factors_copy.csv",
)
```

```python
_FF_MAP: Dict[str, str] = {
    "UMD": "Mom",
    "LTR": "LT_Rev",
    "STR": "ST_Rev",
}

def align_ff_and_compare(
    ff_csv_path: str,
    factors_copy_csv_path: str,
    factor_copy_col: str,
    save_dir: Optional[str] = None,
    save_filtered_csv: Optional[str] = None,
) -> pd.DataFrame:
    """
      1) Load Fama-French CSV with YYYYMM in 'Date'. Map factor via _FF_MAP or infer sole non-Date column.
      2) Coerce FF Date to last NYSE trading weekday via _to_month_last_weekday.
      3) Load factors_copy with 'Date' and factor_copy_col. Coerce Date via _to_month_last_weekday.
      4) Filter FF so start >= factors_copy start. Do not trim factors_copy.
      5) Rename FF factor -> factor_copy_col. Keep [Date, factor_copy_col]. Persist filtered FF aligned CSV.
      6) Merge on Date, compute delta = FF - factors_copy. Summarize min, max, mean.
      7) Plot log-cumulative returns for both series and the delta time series. Persist plots.
    """
    out_dir = Path(save_dir or f"./cleaned_public/ff_{factor_copy_col}_deltas")
    out_dir.mkdir(parents=True, exist_ok=True)

    # ---------- load FF (YYYYMM) ----------
    ff = pd.read_csv(ff_csv_path)
    if "Date" not in ff.columns:
        raise ValueError("Fama-French CSV must contain a 'Date' column with YYYYMM values.")

    ff_cols = [c for c in ff.columns if c != "Date"]
    mapped = _FF_MAP.get(factor_copy_col)
    if mapped and mapped in ff.columns:
        target_ff_col = mapped
    elif len(ff_cols) == 1:
        target_ff_col = ff_cols[0]
    else:
        raise ValueError(
            f"Cannot infer FF factor column. Include '{mapped}' or a single non-Date column."
        )

    ff = ff[["Date", target_ff_col]].copy()
    ff[target_ff_col] = pd.to_numeric(ff[target_ff_col], errors="coerce") / 100.0  # FF is percent

    # Coerce YYYYMM -> first of month -> last NYSE trading day
    ym_ff = pd.to_datetime(
        ff["Date"].astype(str).str[:4] + "-" + ff["Date"].astype(str).str[4:6] + "-01",
        errors="coerce",
    )
    ff["Date"] = _to_month_last_weekday(ym_ff)
    ff = ff.dropna(subset=["Date", target_ff_col]).sort_values("Date").reset_index(drop=True)

    # ---------- load factors_copy ----------
    fc = pd.read_csv(factors_copy_csv_path)
    if "Date" not in fc.columns:
        raise ValueError("factors_copy must contain a 'Date' column.")
    if factor_copy_col not in fc.columns:
        raise ValueError(f"factors_copy must contain '{factor_copy_col}'.")

    fc = fc[["Date", factor_copy_col]].copy()
    fc[factor_copy_col] = pd.to_numeric(fc[factor_copy_col], errors="coerce")
    fc["Date"] = _to_month_last_weekday(fc["Date"])
    fc = fc.dropna(subset=["Date", factor_copy_col]).sort_values("Date").reset_index(drop=True)

    # ---------- align start date to factors_copy ----------
    start_date = fc["Date"].iloc[0]
    ff_aligned = (
        ff.loc[ff["Date"] >= start_date, ["Date", target_ff_col]]
        .sort_values("Date")
        .reset_index(drop=True)
        .copy()
    )

    # ---------- build aligned FF dataset ----------
    ff_aligned = ff_aligned.rename(columns={target_ff_col: factor_copy_col})
    filtered_csv = Path(save_filtered_csv or f"./cleaned_public/FF_aligned_{factor_copy_col}.csv")
    filtered_csv.parent.mkdir(parents=True, exist_ok=True)
    ff_aligned.to_csv(filtered_csv, index=False)

    # ---------- merge and delta ----------
    merged = (
        ff_aligned.merge(fc, on="Date", how="inner", suffixes=("_ff", "_fc"))
        .sort_values("Date")
        .reset_index(drop=True)
    )
    ff_col = f"{factor_copy_col}_ff"
    fc_col = f"{factor_copy_col}_fc"
    merged[[ff_col, fc_col]] = merged[[ff_col, fc_col]].apply(pd.to_numeric, errors="coerce")
    merged = merged.replace([np.inf, -np.inf], np.nan).dropna(subset=[ff_col, fc_col])

    dcol = f"delta_{factor_copy_col}"
    merged[dcol] = merged[ff_col] - merged[fc_col]

    # ---------- summary ----------
    arr = merged[dcol].to_numpy()
    summary = pd.DataFrame([{
        "column": dcol,
        "rows_compared": int(np.sum(~np.isnan(arr))),
        "delta_min": float(np.nanmin(arr)) if arr.size else np.nan,
        "delta_max": float(np.nanmax(arr)) if arr.size else np.nan,
        "delta_mean": float(np.nanmean(arr)) if arr.size else np.nan,
    }])

    # ---------- plots ----------
    dfp = merged[["Date", ff_col, fc_col]].copy()
    for c in [ff_col, fc_col]:
        r = pd.to_numeric(dfp[c], errors="coerce").mask(dfp[c] <= -1.0)
        dfp[f"log_cum_{c}"] = np.log1p(r).cumsum()

    plt.figure()
    plt.plot(dfp["Date"], dfp[f"log_cum_{ff_col}"], label=f"{factor_copy_col} (FF)")
    plt.plot(dfp["Date"], dfp[f"log_cum_{fc_col}"], label=f"{factor_copy_col} (factors_copy)")
    plt.title(f"Log cumulative return: {factor_copy_col}")
    plt.xlabel("Date")
    plt.ylabel("log cum return")
    plt.legend()
    plt.tight_layout()
    plt.savefig(out_dir / f"log_cum_{factor_copy_col}.png")
    plt.close()

    plt.figure()
    plt.plot(merged["Date"], merged[dcol])
    plt.title(f"Delta over time: {factor_copy_col} (FF - factors_copy)")
    plt.xlabel("Date")
    plt.ylabel("Delta")
    plt.tight_layout()
    plt.savefig(out_dir / f"delta_{factor_copy_col}.png")
    plt.close()

    return summary

align_ff_and_compare(
    ff_csv_path="./public_source/F-F_Momentum_Factor.csv",
    factors_copy_csv_path="factors_copy.csv",
    factor_copy_col="UMD"
)
```

```python
def align_q_factors_and_compare(
    q_csv_path: str,
    factors_copy_csv_path: str,
    save_dir: Optional[str] = None,
    save_filtered_csv: Optional[str] = None,
) -> pd.DataFrame:
    """
    1) Load q-factors with 'year' and 'month'. Build Date as last trading day of month.
    2) Load factors_copy. Coerce Date to last trading day of month.
    3) Map: R_IA->HXZ_IA, R_ROE->HXZ_ROE in q-factors, then keep only Date, HXZ_IA, HXZ_ROE.
    4) Filter q-factors so its start date >= factors_copy's start date.
    5) Merge on Date and compute deltas: q - factors_copy for HXZ_IA and HXZ_ROE.
    6) Plot log cumulative returns for HXZ_IA and HXZ_ROE from both datasets.
    7) Persist plots and filtered q-factors aligned CSV. Return delta summary.
    """
    out_dir = Path(save_dir or "./cleaned_public/q_factors_deltas")
    out_dir.mkdir(parents=True, exist_ok=True)

    # --- Load q-factors ---
    q = pd.read_csv(q_csv_path)
    year_col = next((c for c in q.columns if c.lower() == "year"), None)
    month_col = next((c for c in q.columns if c.lower() == "month"), None)
    if year_col is None or month_col is None:
        raise ValueError("q-factors CSV must have 'year' and 'month' columns.")

    ym = pd.to_datetime(
        q[year_col].astype(int).astype(str).str.zfill(4) + "-" + q[month_col].astype(int).astype(str).str.zfill(2) + "-01",
        errors="coerce",
    )
    q["Date"] = _to_month_last_weekday(ym)

    rename_map = {"R_IA": "HXZ_IA", "R_ROE": "HXZ_ROE"}
    for src, _ in rename_map.items():
        if src not in q.columns:
            raise ValueError(f"q-factors missing required column '{src}'.")
    q = q.rename(columns=rename_map)
    q[["HXZ_IA", "HXZ_ROE"]] = q[["HXZ_IA", "HXZ_ROE"]].apply(pd.to_numeric, errors="coerce") / 100.0
    q = q[["Date", "HXZ_IA", "HXZ_ROE"]].copy()

    # --- Load factors_copy ---
    fc = pd.read_csv(factors_copy_csv_path)
    date_col = next((c for c in fc.columns if c.lower() == "date"), None)
    if date_col is None:
        raise ValueError("factors_copy must contain a 'Date' column.")
    if date_col != "Date":
        fc = fc.rename(columns={date_col: "Date"})
    for c in ["HXZ_IA", "HXZ_ROE"]:
        if c not in fc.columns:
            raise ValueError(f"factors_copy missing required column '{c}'.")
    fc["Date"] = _to_month_last_weekday(fc["Date"])
    fc = fc.dropna(subset=["Date"]).sort_values("Date").reset_index(drop=True)

    # --- Align start date ---
    start_date = fc["Date"].iloc[0]
    q_aligned = q[q["Date"] >= start_date].dropna(subset=["Date"]).sort_values("Date").reset_index(drop=True)

    # Persist filtered q-factors aligned CSV
    filtered_csv = Path(save_filtered_csv or "./cleaned_public/QFactors_aligned.csv")
    filtered_csv.parent.mkdir(parents=True, exist_ok=True)
    q_aligned.to_csv(filtered_csv, index=False)

    # --- Merge and deltas ---
    merged = (
        q_aligned.merge(fc[["Date", "HXZ_IA", "HXZ_ROE"]], on="Date", how="inner", suffixes=("_q", "_fc"))
        .sort_values("Date")
        .reset_index(drop=True)
    )
    for c in ["HXZ_IA_q", "HXZ_IA_fc", "HXZ_ROE_q", "HXZ_ROE_fc"]:
        merged[c] = pd.to_numeric(merged[c], errors="coerce")
    merged = merged.replace([np.inf, -np.inf], np.nan).dropna(subset=["HXZ_IA_q", "HXZ_IA_fc", "HXZ_ROE_q", "HXZ_ROE_fc"])

    merged["delta_HXZ_IA"] = merged["HXZ_IA_q"] - merged["HXZ_IA_fc"]
    merged["delta_HXZ_ROE"] = merged["HXZ_ROE_q"] - merged["HXZ_ROE_fc"]

    # --- Summary ---
    def _summ(col: str) -> dict:
        arr = merged[col].to_numpy()
        return {
            "column": col,
            "rows_compared": int(np.sum(~np.isnan(arr))),
            "delta_min": float(np.nanmin(arr)),
            "delta_max": float(np.nanmax(arr)),
            "delta_mean": float(np.nanmean(arr)),
        }

    summary = pd.DataFrame([_summ("delta_HXZ_IA"), _summ("delta_HXZ_ROE")])

    # --- Plot log cumulative returns ---
    def _plot_log_cum(pair: Tuple[str, str], title: str, out_name: str, out_dir: Path) -> None:
        a, b = pair
        dfp = merged[["Date", a, b]].copy()
        for c in [a, b]:
            dfp = dfp[dfp[c] > -1.0]
        dfp["log_cum_a"] = np.log1p(dfp[a]).cumsum()
        dfp["log_cum_b"] = np.log1p(dfp[b]).cumsum()

        plt.figure()
        plt.plot(dfp["Date"], dfp["log_cum_a"], label=f"{a}")
        plt.plot(dfp["Date"], dfp["log_cum_b"], label=f"{b}")
        plt.title(title)
        plt.xlabel("Date")
        plt.ylabel("log cum return")
        plt.legend()
        plt.tight_layout()
        plt.savefig(out_dir / out_name)
        plt.close()

    out_dir = Path(save_dir or "./cleaned_public/q_factors_deltas")
    out_dir.mkdir(parents=True, exist_ok=True)
    _plot_log_cum(("HXZ_IA_q", "HXZ_IA_fc"), "Log cumulative return: HXZ_IA (q vs factors_copy)", "hxz_ia_log_cum.png", out_dir)
    _plot_log_cum(("HXZ_ROE_q", "HXZ_ROE_fc"), "Log cumulative return: HXZ_ROE (q vs factors_copy)", "hxz_roe_log_cum.png", out_dir)

    plt.figure()
    plt.plot(merged["Date"], merged["delta_HXZ_IA"])
    plt.title("Delta over time: HXZ_IA (q - factors_copy)")
    plt.xlabel("Date")
    plt.ylabel("Delta")
    plt.tight_layout()
    plt.savefig(out_dir / "delta_HXZ_IA.png")
    plt.close()

    plt.figure()
    plt.plot(merged["Date"], merged["delta_HXZ_ROE"])
    plt.title("Delta over time: HXZ_ROE (q - factors_copy)")
    plt.xlabel("Date")
    plt.ylabel("Delta")
    plt.tight_layout()
    plt.savefig(out_dir / "delta_HXZ_ROE.png")
    plt.close()

    # merged.to_csv(out_dir / "q_vs_factors_copy_merged.csv", index=False)
    return summary

align_q_factors_and_compare(
    q_csv_path="./public_source/q5_factors_monthly_2024.csv",
    factors_copy_csv_path="factors_copy.csv",
)
```

```python
def align_intermediary_and_compare(
    intermediary_csv_path: str,
    factors_copy_csv_path: str,
    save_dir: Optional[str] = None,
    save_filtered_csv: Optional[str] = None,
) -> pd.DataFrame:
    """
    1) Load intermediary factors with a YYYYMM column. Rename it to 'Date' and coerce to last trading weekday via _to_month_last_weekday.
    2) Load factors_copy. Coerce its 'Date' to last trading weekday via _to_month_last_weekday.
    3) Map: intermediary_capital_risk_factor -> Intermediary in intermediary data. Keep only Date, Intermediary.
    4) Filter intermediary so its start date >= factors_copy's start date.
    5) Merge on Date and compute delta = intermediary_csv - factors_copy for Intermediary.
    6) Plot log cumulative returns for Intermediary from both datasets and the delta series. Persist outputs.
    7) Persist the filtered intermediary aligned CSV. Return delta summary (min, max, mean, rows).
    """
    out_dir = Path(save_dir or "./cleaned_public/intermediary_deltas")
    out_dir.mkdir(parents=True, exist_ok=True)

    # ---------- load intermediary ----------
    inter = pd.read_csv(intermediary_csv_path)

    # Date from exact 'yyyymm'
    ym = pd.to_datetime(
        inter["yyyymm"].astype(str).str[:4] + "-" + inter["yyyymm"].astype(str).str[4:6] + "-01",
        errors="coerce",
    )
    inter = inter.rename(columns={"yyyymm": "Date", "intermediary_capital_risk_factor": "Intermediary"})
    inter["Date"] = _to_month_last_weekday(ym)
    inter = inter[["Date", "Intermediary"]].copy()
    inter["Intermediary"] = pd.to_numeric(inter["Intermediary"], errors="coerce")
    inter = inter.dropna(subset=["Date"]).sort_values("Date").reset_index(drop=True)

    # ---------- load factors_copy ----------
    fc = pd.read_csv(factors_copy_csv_path)
    if "Date" not in fc.columns:
        raise ValueError("factors_copy must contain a 'Date' column.")
    if "Intermediary" not in fc.columns:
        raise ValueError("factors_copy must contain an 'Intermediary' column.")

    fc["Date"] = _to_month_last_weekday(fc["Date"])
    fc["Intermediary"] = pd.to_numeric(fc["Intermediary"], errors="coerce")
    fc = fc.dropna(subset=["Date"]).sort_values("Date").reset_index(drop=True)

    # ---------- align start date ----------
    start_date = fc["Date"].iloc[0]
    inter_aligned = inter[inter["Date"] >= start_date].dropna(subset=["Date"]).sort_values("Date").reset_index(drop=True)

    # persist filtered intermediary aligned CSV
    filtered_csv = Path(save_filtered_csv or "./cleaned_public/Intermediary_aligned.csv")
    filtered_csv.parent.mkdir(parents=True, exist_ok=True)
    inter_aligned.to_csv(filtered_csv, index=False)

    # ---------- merge and delta ----------
    merged = (
        inter_aligned.merge(fc[["Date", "Intermediary"]], on="Date", how="inner", suffixes=("_inter", "_fc"))
        .sort_values("Date")
        .reset_index(drop=True)
    )
    merged[["Intermediary_inter", "Intermediary_fc"]] = merged[["Intermediary_inter", "Intermediary_fc"]].apply(
        pd.to_numeric, errors="coerce"
    )
    merged = merged.replace([np.inf, -np.inf], np.nan).dropna(subset=["Intermediary_inter", "Intermediary_fc"])

    merged["delta_Intermediary"] = merged["Intermediary_inter"] - merged["Intermediary_fc"]

    # ---------- summary ----------
    arr = merged["delta_Intermediary"].to_numpy()
    summary = pd.DataFrame(
        [{
            "column": "delta_Intermediary",
            "rows_compared": int(np.sum(~np.isnan(arr))),
            "delta_min": float(np.nanmin(arr)) if arr.size else np.nan,
            "delta_max": float(np.nanmax(arr)) if arr.size else np.nan,
            "delta_mean": float(np.nanmean(arr)) if arr.size else np.nan,
        }]
    )

    # ---------- plots ----------
    # log cumulative returns
    dfp = merged[["Date", "Intermediary_inter", "Intermediary_fc"]].copy()
    for c in ["Intermediary_inter", "Intermediary_fc"]:
        r = pd.to_numeric(dfp[c], errors="coerce")
        r = r.mask(r <= -1.0)  
        dfp[f"log_cum_{c}"] = np.log1p(r).cumsum()

    plt.figure()
    plt.plot(dfp["Date"], dfp["log_cum_Intermediary_inter"], label="Intermediary (intermediary_csv)")
    plt.plot(dfp["Date"], dfp["log_cum_Intermediary_fc"], label="Intermediary (factors_copy)")
    plt.title("Log cumulative return: Intermediary")
    plt.xlabel("Date")
    plt.ylabel("log cum return")
    plt.legend()
    plt.tight_layout()
    plt.savefig(out_dir / "intermediary_log_cum.png")
    plt.close()

    # delta time series
    plt.figure()
    plt.plot(merged["Date"], merged["delta_Intermediary"])
    plt.title("Delta over time: Intermediary (intermediary_csv - factors_copy)")
    plt.xlabel("Date")
    plt.ylabel("Delta")
    plt.tight_layout()
    plt.savefig(out_dir / "delta_Intermediary.png")
    plt.close()

    # audit
    # merged.to_csv(out_dir / "intermediary_vs_factors_copy_merged.csv", index=False)

    return summary

align_intermediary_and_compare(
    intermediary_csv_path="./public_source/He_Kelly_Manela_Factors_monthly_250627.csv",
    factors_copy_csv_path="factors_copy.csv",
)
```

```python
def align_aqr_and_compare(
    aqr_xlsx_path: str,
    factors_copy_csv_path: str,
    factor_copy_col: str,
    save_dir: Optional[str] = None,
    save_filtered_csv: Optional[str] = None,
) -> pd.DataFrame:
    """
      1) Load AQR (sheet 0, skip 18). Keep DATE and USA. DATE -> datetime, coerce to last trading weekday. USA -> numeric /100.
      2) Load factors_copy. Require 'Date' and factor_copy_col. Coerce Date to last trading weekday.
      3) Filter AQR so Date >= factors_copy first Date. Do not trim factors_copy.
      4) Rename AQR USA -> factor_copy_col. Keep [Date, factor_copy_col].
      5) Merge on Date and compute delta = AQR - factors_copy.
      6) Persist: aligned AQR CSV, log-cum plots for both series, delta plot.
      7) Return summary DataFrame with rows, min, max, mean of delta.
    """
    out_dir = Path(save_dir or f"./cleaned_public/aqr_{factor_copy_col}_deltas")
    out_dir.mkdir(parents=True, exist_ok=True)

    # ---------- load AQR ----------
    aqr = pd.read_excel(
        aqr_xlsx_path,
        sheet_name=0,
        skiprows=18,
        engine="openpyxl",
    )
    if "DATE" not in aqr.columns:
        raise ValueError("AQR sheet must contain a 'DATE' column on the first sheet starting at row 19.")
    if "USA" not in aqr.columns:
        raise ValueError("AQR sheet must contain a 'USA' column.")

    aqr = aqr.rename(columns={"DATE": "Date"})
    aqr["Date"] = pd.to_datetime(aqr["Date"], errors="coerce")
    # normalize to last trading weekday to match your pipeline
    aqr["Date"] = _to_month_last_weekday(aqr["Date"])
    aqr["USA"] = pd.to_numeric(aqr["USA"], errors="coerce")
    aqr = aqr.dropna(subset=["Date", "USA"]).sort_values("Date").reset_index(drop=True)

    # ---------- load factors_copy ----------
    fc = pd.read_csv(factors_copy_csv_path)
    if "Date" not in fc.columns:
        raise ValueError("factors_copy must contain a 'Date' column.")
    if factor_copy_col not in fc.columns:
        raise ValueError(f"factors_copy must contain '{factor_copy_col}'.")

    fc["Date"] = _to_month_last_weekday(fc["Date"])
    fc[factor_copy_col] = pd.to_numeric(fc[factor_copy_col], errors="coerce")
    fc = fc.dropna(subset=["Date", factor_copy_col]).sort_values("Date").reset_index(drop=True)

    # ---------- align start date to factors_copy ----------
    start_date = fc["Date"].iloc[0]
    aqr_aligned = (
        aqr.loc[aqr["Date"] >= start_date, ["Date", "USA"]]
        .sort_values("Date")
        .reset_index(drop=True)
        .copy()
    )

    # ---------- build aligned AQR dataset with target name ----------
    aqr_aligned = aqr_aligned.rename(columns={"USA": factor_copy_col})
    # persist filtered AQR aligned CSV
    filtered_csv = Path(save_filtered_csv or f"./cleaned_public/AQR_aligned_{factor_copy_col}.csv")
    filtered_csv.parent.mkdir(parents=True, exist_ok=True)
    aqr_aligned.to_csv(filtered_csv, index=False)

    # ---------- merge on Date and delta ----------
    merged = (
        aqr_aligned.merge(fc[["Date", factor_copy_col]], on="Date", how="inner", suffixes=("_aqr", "_fc"))
        .sort_values("Date")
        .reset_index(drop=True)
    )
    aqr_col = f"{factor_copy_col}_aqr"
    fc_col = f"{factor_copy_col}_fc"
    merged[[aqr_col, fc_col]] = merged[[aqr_col, fc_col]].apply(pd.to_numeric, errors="coerce")
    merged = merged.replace([np.inf, -np.inf], np.nan).dropna(subset=[aqr_col, fc_col])

    dcol = f"delta_{factor_copy_col}"
    merged[dcol] = merged[aqr_col] - merged[fc_col]

    # ---------- summary ----------
    arr = merged[dcol].to_numpy()
    summary = pd.DataFrame([{
        "column": dcol,
        "rows_compared": int(np.sum(~np.isnan(arr))),
        "delta_min": float(np.nanmin(arr)) if arr.size else np.nan,
        "delta_max": float(np.nanmax(arr)) if arr.size else np.nan,
        "delta_mean": float(np.nanmean(arr)) if arr.size else np.nan,
    }])

    # ---------- plots ----------
    dfp = merged[["Date", aqr_col, fc_col]].copy()
    for c in [aqr_col, fc_col]:
        r = pd.to_numeric(dfp[c], errors="coerce")
        r = r.mask(r <= -1.0)
        dfp[f"log_cum_{c}"] = np.log1p(r).cumsum()

    plt.figure()
    plt.plot(dfp["Date"], dfp[f"log_cum_{aqr_col}"], label=f"{factor_copy_col} (AQR)")
    plt.plot(dfp["Date"], dfp[f"log_cum_{fc_col}"], label=f"{factor_copy_col} (factors_copy)")
    plt.title(f"Log cumulative return: {factor_copy_col}")
    plt.xlabel("Date")
    plt.ylabel("log cum return")
    plt.legend()
    plt.tight_layout()
    plt.savefig(out_dir / f"log_cum_{factor_copy_col}.png")
    plt.close()

    plt.figure()
    plt.plot(merged["Date"], merged[dcol])
    plt.title(f"Delta over time: {factor_copy_col} (AQR - factors_copy)")
    plt.xlabel("Date")
    plt.ylabel("Delta")
    plt.tight_layout()
    plt.savefig(out_dir / f"delta_{factor_copy_col}.png")
    plt.close()

    return summary

align_aqr_and_compare(
    aqr_xlsx_path="./public_source/Betting Against Beta Equity Factors Monthly.xlsx",
    factors_copy_csv_path="factors_copy.csv",
    factor_copy_col="BAB",
)
```

```python
def merge_factor_files(csv_paths: Sequence[str | Path],
                       save_path: str | Path = "public_sourced_factors.csv"
                      ) -> pd.DataFrame:
    """
    Merge multiple monthly factor CSVs on 'Date' and persist.
      - Inner-joins on 'Date' to keep only months present in all files.
      - On duplicate column names, suffix with the file stem.
      - Writes CSV with ISO dates.
    """
    if not csv_paths:
        raise ValueError("Provide at least one CSV path.")

    dfs: list[pd.DataFrame] = []
    spans: list[tuple[pd.Timestamp, pd.Timestamp]] = []

    for p in csv_paths:
        p = Path(p)
        df = pd.read_csv(p)
        if "Date" not in df.columns:
            raise ValueError(f"{p} is missing 'Date' column.")
        df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
        df = df.dropna(subset=["Date"]).sort_values("Date")
        df = df[~df["Date"].duplicated(keep="last")]  

        spans.append((df["Date"].min(), df["Date"].max()))

        cols = [c for c in df.columns if c != "Date"]
        # Disambiguate duplicate factor names across files
        rename = {}
        existing_cols = set().union(*(set(d.columns) for d in dfs)) if dfs else set()
        for c in cols:
            if c in existing_cols:
                rename[c] = f"{c}__{p.stem}"
        if rename:
            df = df.rename(columns=rename)
            cols = [rename.get(c, c) for c in cols]

        dfs.append(df[["Date"] + cols])

    combined = reduce(lambda l, r: pd.merge(l, r, on="Date", how="inner"), dfs)

    # Enforce earliest end date explicitly, and shared start if inputs differ
    latest_start = max(s for s, _ in spans)
    earliest_end = min(e for _, e in spans)
    combined = (
        combined[(combined["Date"] >= latest_start) & (combined["Date"] <= earliest_end)]
        .sort_values("Date")
        .reset_index(drop=True)
    )

    save_path = Path(save_path)
    combined.to_csv(save_path, index=False, date_format="%Y-%m-%d")
    return combined

merge_factor_files([
    "./cleaned_public/FF5_aligned.csv",
    "./cleaned_public/FF_aligned_UMD.csv",
    "./cleaned_public/FF_aligned_LTR.csv",
    "./cleaned_public/FF_aligned_STR.csv",
    "./cleaned_public/LIQ_aligned.csv",
    "./cleaned_public/QFactors_aligned.csv",
    "./cleaned_public/Intermediary_aligned.csv",
    "./cleaned_public/AQR_aligned_BAB.csv",
    "./cleaned_public/AQR_aligned_QMJ.csv",
    "./cleaned_public/AQR_aligned_HML_Devil.csv",
], save_path="public_sourced_factors.csv")
```
